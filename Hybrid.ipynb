## RUMUS HYBRID

```python
def HybridFiltering(PrediksiUser, PrediksiItem, gamma=0.7):
    # Menghitung prediksi hybrid
    rumusHybrid = gamma * PrediksiUser + (1 - gamma) * PrediksiItem
    # mengembalikan hasil prediksi hybrid
    return rumusHybrid
```

## LOOP VARIASI HYBRID

```python
def loopHybridParameterK(user_path, item_path, hybrid_output_path):
    # loop untuk tetangga kUser
    for kUser in variasiParameterKUser:
        #loop untuk tetangga kItem
        for kItem in variasiParameterKItem2:
            # loop untuk gamma
            for gamma in variasiParameterHybrid:
                # loop untuk fold 1-5
                for fold in range(1, 6):
                    # Path file input user & item
                    path_user = os.path.join(user_path, str(fold), f"{kUser}.joblib")
                    path_item = os.path.join(item_path, str(fold), f"{kItem}.joblib")

                    if not os.path.exists(path_user) or not os.path.exists(path_item):
                        print(f"❌ File tidak ditemukan: {path_user} atau {path_item}")
                        continue

                    # Load prediksi
                    PrediksiUser = joblib.load(path_user)
                    PrediksiItem = joblib.load(path_item)

                    # Hitung hybrid
                    CutHybrid = HybridFiltering(PrediksiUser, PrediksiItem, gamma)
                    # CutHybrid = CutHybrid.loc[1:463, 1:1592]

                    # Folder output per fold
                    folder_output = os.path.join(hybrid_output_path, str(fold))
                    os.makedirs(folder_output, exist_ok=True)

                    # Simpan hasil
                    nama_file = f"{kUser}_{kItem}_{gamma}.joblib"
                    path_simpan = os.path.join(folder_output, nama_file)
                    joblib.dump(CutHybrid, path_simpan)

                    print(f"✅ Disimpan: {path_simpan}")
```

```python
def TopN_fromPrediksi(user_id, rating_matrix, prediksi_matrix, N=5):
    # Konversi ke numpy array
    user_pred = prediksi_matrix.iloc[user_id].to_numpy()
    user_actual = rating_matrix.iloc[user_id].to_numpy()
    print(f" user actual {len(user_actual)}")
    # Item yang belum dirating oleh user
    unrated_items = np.where(user_actual == 0)[0]
    print(f"Jumlah item yang BELUM dirating: {len(unrated_items)}")
    # Ambil prediksi hanya untuk item yang belum dirating
    pred_unrated = user_pred[unrated_items]
    # Urutkan descending berdasarkan prediksi
    topN_indices = np.argsort(-pred_unrated)[:N]
    # Ambil item ID dari unrated_items
    topN_items = unrated_items[topN_indices]
    # Buat list (item_id, skor prediksi)
    # listItem = [(item_id, user_pred[item_id]) for item_id in topN_items]
    # print(f'list item prediksi {listItem}')

    return topN_items
```

```python
def KumpulkanTopN2(rating_matrix, prediksi_matrix, N=5):
    # membuat list kosong
    semua_topN = []
    # Loop untuk setiap user berdasarkan jumlah baris / user
    for user_id in range(len(rating_matrix)):
        # Ambil top N item untuk user_id
        topN_items = TopN_fromPrediksi(user_id, rating_matrix, prediksi_matrix, N)
        # Simpan hasil ke dalam list
        semua_topN.append(topN_items.tolist())
    return semua_topN
```
