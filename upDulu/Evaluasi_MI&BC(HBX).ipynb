```python
from tqdm.auto import tqdm
from natsort import natsorted
```

```python
import pandas as pd
import numpy as np
import os
import joblib
# from sklearn.metrics import adjusted_mutual_info_score
```

# Menghitung Akurasi Fold

```python
def calculateAkurasi(path):
    evaluasi = {}
    for x, linkpathFold in enumerate(tqdm(natsorted(os.listdir(path)))):
        for linkpath in tqdm(natsorted(os.listdir(os.path.join(path, linkpathFold, 'Evaluasi')))):
            ev = np.array(joblib.load(os.path.join(path, linkpathFold, "Evaluasi", linkpath)))
            meanEv = (ev[:,4,19]).mean(axis=0)
            pathName = linkpath.split(f'_{x+1}Fold.ev')[0]
            if pathName in evaluasi:
                evaluasi[pathName].append(meanEv)
            else:
                evaluasi[pathName] = [meanEv]
    return evaluasi
```

# Menghitung Komputasi

```python
def calculateKomputasi(path):
    komputasi = {}
    for x, linkpathFold in enumerate(tqdm(natsorted(os.listdir(path)))):
        for linkpath in tqdm(natsorted(os.listdir(os.path.join(path, linkpathFold, 'Waktu')))):
            waktu = np.array(joblib.load(os.path.join(path, linkpathFold, "Waktu", linkpath)))
            pathName = linkpath.split(f'_{x+1}Fold.time')[0]
            if pathName in komputasi:
                komputasi[pathName].append(waktu)
            else:
                komputasi[pathName] = [waktu]
    return komputasi
```

# Parameter skenario path

```python
HB = "HBX"
skenario = 'MI&BC'

pathRoot = '/content/drive/MyDrive/@Ahmad Farisul Haq (200411100171)_Skripsi/Kode'
path = os.path.join(pathRoot, f'Eksperimen/Request Bu IFA/{HB}/{skenario}')


pathEvaluasiU = os.path.join(path, 'tetanggaU')
pathEvaluasiI = os.path.join(path, 'tetanggaI')
pathEvaluasiR = os.path.join(path, 'R')

pathKomputasiU = os.path.join(path, 'tetanggaU')
pathKomputasiI = os.path.join(path, 'tetanggaI')
pathKomputasiR = os.path.join(path, 'R')
```

# Skenario

## UCF

### Parameter terbaik tetangga user

```python
evaluasiU = calculateAkurasi(pathEvaluasiU)
```

```python
df_evaluasiU = pd.DataFrame(evaluasiU, index=[f'{i}Fold' for i in range(1,6)]).T
```

```python
df_evaluasiU['Rata-Rata'] = df_evaluasiU.mean(axis=1)
df_evaluasiU
```

```python
df_evaluasiU.to_excel(f'{skenario}_UCF-{HB}.xlsx')
```

### Waktu Komputasi

```python
komputasiU = calculateKomputasi(pathKomputasiU)
```

```python
df_komputasiU = pd.DataFrame(komputasiU, index=[f'{i}Fold' for i in range(1,6)]).T
df_komputasiU
```
