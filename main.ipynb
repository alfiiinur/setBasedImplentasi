```python
def simJaccardUser(RatingMatriks, penggunaU, penggunaV):
    """
    Fungsi untuk menghitung similarity Jaccard antara dua pengguna
    """
    # mendapatkan item yang diberi rating oleh penggunaU dan penggunaV
    itemRatedPenggunaU = set(np.where(RatingMatriks.loc[penggunaU, :] != 0)[0])  # Indeks item yang diberi rating oleh penggunaU
    itemRatedPenggunaV = set(np.where(RatingMatriks.loc[penggunaV, :] != 0)[0])  # Indeks item yang diberi rating oleh penggunaV
    
    # Menghitung irisan (intersection) dari item yang diberi rating oleh kedua pengguna
    intersection = len(itemRatedPenggunaU.intersection(itemRatedPenggunaV))

    # Menghitung gabungan (union) dari item yang diberi rating oleh kedua pengguna
    union = len(itemRatedPenggunaU.union(itemRatedPenggunaV))

    # Menghitung similarity Jaccard
    rumusJaccardSim = intersection / union if union != 0 else 0
    return rumusJaccardSim

    
```

```python
def similarityJaccardUser(RatingMatriks):
    """
    Fungsi untuk menghitung similarity Jaccard antara semua pengguna.
    Mengembalikan DataFrame di mana diagonalnya adalah 1 (similarity dengan dirinya sendiri adalah 1).
    
    Argumen:
    - matriksRating (pd.DataFrame): Matriks pengguna-item yang menunjukkan rating (0 jika tidak ada rating).
    
    Mengembalikan:
    - pd.DataFrame: Matriks similarity Jaccard antara semua pengguna.
    """
    # Mendapatkan jumlah pengguna
    jumlahPengguna = RatingMatriks.shape[0]

    # Membuat matriks similarity kosong
    matriksSimilarity = np.zeros((jumlahPengguna, jumlahPengguna))

    # Menghitung similarity Jaccard untuk setiap pasangan pengguna
    for i in range(jumlahPengguna):
        for j in range(jumlahPengguna):
            if i == j:
                matriksSimilarity[i][j] = 1  # Jika pengguna yang dibandingkan sama, similarity adalah 1
            elif i != j:
                matriksSimilarity[i][j] = simJaccardUser(RatingMatriks, RatingMatriks.index[i], RatingMatriks.index[j])

    # Mengembalikan hasil dalam bentuk DataFrame
    dfMatriksSimilarity = pd.DataFrame(matriksSimilarity, index=RatingMatriks.index, columns=RatingMatriks.index)
    
    return dfMatriksSimilarity
```

```python
# USER-BASED JACCARD

def jaccardSimilarityUser(RatingMatriks):
    """
    Menghitung Jaccard similarity antara pengguna berdasarkan RatingMatriks.

    Args:
    - RatingMatriks (pd.DataFrame): Matriks pengguna-film yang menunjukkan rating (0 jika tidak ada rating).

    Returns:
    - pd.DataFrame: Matriks Jaccard similarity antara semua pengguna.
    """
    # Convert the RatingMatriks to a numpy array
    npRatingMatriks = np.array(RatingMatriks)
    
    # Initialize an empty matrix to store the Jaccard similarities
    jaccardSimilarityUser = np.zeros((npRatingMatriks.shape[0], npRatingMatriks.shape[0]))  # Shape should be (users, users)
    
    # Iterate through all pairs of users (rows)
    for i in range(npRatingMatriks.shape[0]):  # Loop through rows (users)
        for j in range(npRatingMatriks.shape[0]):  # Loop through rows (users)
            if i == j:
                jaccardSimilarityUser[i][j] = 1  # Jaccard similarity of a user with itself is 1
            else:
                # Find items rated by both users (intersection)
                rated_by_user1 = set(np.where(npRatingMatriks[i, :] != 0)[0])
                rated_by_user2 = set(np.where(npRatingMatriks[j, :] != 0)[0])
                
                # Calculate intersection (items rated by both users)
                intersection = len(rated_by_user1.intersection(rated_by_user2))
                
                # Calculate union (all distinct items rated by either user)
                union = len(rated_by_user1.union(rated_by_user2))
                
                # Compute the Jaccard similarity
                if union != 0:
                    jaccardSimilarityUser[i][j] = intersection / union
                else:
                    jaccardSimilarityUser[i][j] = 0  # If union is 0 (no ratings in common), similarity is 0
    
    # Convert the result back to a DataFrame with proper indices and columns (users)
    dfJaccardSimilarityUser = pd.DataFrame(jaccardSimilarityUser, index=RatingMatriks.index, columns=RatingMatriks.index)
    
    return dfJaccardSimilarityUser
```

```python
def simJaccardItem(RatingMatriks, itemI, itemJ) :
    """
    Fungsi untuk menghitung similarity Jaccard antara dua item
    """
    # mendapatkan pengguna yang memberi rating pada itemI dan itemJ
    userRatedItemI = set(np.where(RatingMatriks.loc[:, itemI] != 0)[0])  # Indeks pengguna yang memberi rating pada itemI
    userRatedItemJ = set(np.where(RatingMatriks.loc[:, itemJ] != 0)[0])  # Indeks pengguna yang memberi rating pada itemJ
    
    # Menghitung irisan (intersection) dari pengguna yang memberi rating pada kedua item
    intersection = len(userRatedItemI.intersection(userRatedItemJ))

    # Menghitung gabungan (union) dari pengguna yang memberi rating pada kedua item
    union = len(userRatedItemI.union(userRatedItemJ))

    # rumus jaccard item
    rumusJaccardSimItem = intersection / union if union != 0 else 0
    return rumusJaccardSimItem
    
```

```python
def similarityJaccardItem(RatingMatriks):
    """
    Fungsi untuk menghitung similarity Jaccard antara semua item.
    Mengembalikan DataFrame di mana diagonalnya adalah 1 (similarity dengan dirinya sendiri adalah 1).
    
    Argumen:
    - matriksRating (pd.DataFrame): Matriks pengguna-item yang menunjukkan rating (0 jika tidak ada rating).
    
    Mengembalikan:
    - pd.DataFrame: Matriks similarity Jaccard antara semua item.
    """
    # Mendapatkan jumlah item
    jumlahItem = RatingMatriks.shape[1]

    # Membuat matriks similarity kosong
    matriksSimilarity = np.zeros((jumlahItem, jumlahItem))

    # Menghitung similarity Jaccard untuk setiap pasangan item
    for i in range(jumlahItem):
        for j in range(jumlahItem):
            if i == j:
                matriksSimilarity[i][j] = 1  # Jika item yang dibandingkan sama, similarity adalah 1
            elif i != j:
                matriksSimilarity[i][j] = simJaccardItem(RatingMatriks, RatingMatriks.columns[i], RatingMatriks.columns[j])

    # Mengembalikan hasil dalam bentuk DataFrame
    dfMatriksSimilarity = pd.DataFrame(matriksSimilarity, index=RatingMatriks.columns, columns=RatingMatriks.columns)
    
    return dfMatriksSimilarity
```

#### USER-BASED

```python
def simRJUser(RatingMatriks, penggunaU, penggunaV):
    # mendapatkan item yang diberi RatingMatriks oleh penggunaU dan penggunaV
    itemRatedPenggunaU = set(np.where(RatingMatriks.loc[penggunaU, :] != 0)[0])
    itemRatedPenggunav = set(np.where(RatingMatriks.loc[penggunaV, :] != 0)[0])
    
    # mencari irisan dari item yang diberi rating oleh penggunaU dan penggunaV
    intersection = len(itemRatedPenggunaU.intersection(itemRatedPenggunav))
    iPenggunaU = len(itemRatedPenggunaU)
    iPenggunaV = len(itemRatedPenggunav)

    # menghitung similarity
    if intersection != 0:
        rumusSimRJUser = 1/(1+(1/intersection)+(iPenggunaU/1+iPenggunaU)+(iPenggunaV/1+iPenggunaV))
    elif intersection == 0:
        rumusSimRJUser = 0
    
    return rumusSimRJUser
```

```python
def similarityRJUser(RatingMatriks):
    jumlahPengguna = RatingMatriks.shape[0]

    # membuat matriks kosong untuk menyimpan hasil similarity
    simMatriks = np.zeros((jumlahPengguna, jumlahPengguna))
    # menghitung similarity antar pengguna
    for i in range(jumlahPengguna):
        for j in range(jumlahPengguna):
            if i == j:
                simMatriks[i][j] = simRJUser(RatingMatriks, RatingMatriks.index[i], RatingMatriks.index[j])
            elif i != j :
                simMatriks[i][j] = simRJUser(RatingMatriks, RatingMatriks.index[i], RatingMatriks.index[j])
    
    # mengubah matriks similarity menjadi dataframe
    simMatriksDf = pd.DataFrame(simMatriks, index=RatingMatriks.index, columns=RatingMatriks.index)

    return simMatriksDf
```

```python
# prediksi user-based

def prediksiCF(RatingMatriks, similarityFunction, mean, meanCen, user=2, item=2, k=2, jenis="userBased"):

    if jenis == "userBased":
        # mendapatkan item yang diberi rating oleh pengguna
        ratingMatriks = RatingMatriks.loc[:, item].to_numpy()
        # print(ratingMatriks)
        # mendapatan mean centered
        meanCentered = meanCen.loc[:, item].to_numpy()
        # print(meanCentered)
        # mendapatkan similarity
        similarity = similarityFunction.loc[user, :].to_numpy()
        # print(similarity)

    if jenis == "itemBased":
        # mendapatkan item yang diberi rating oleh pengguna
        ratingMatriks = RatingMatriks.loc[user, :].to_numpy()
        # print(ratingMatriks)
        # mendapatan mean centered
        meanCentered = meanCen.loc[user, :].to_numpy()
        # print(meanCentered)
        # mendapatkan similarity
        similarity = similarityFunction.loc[item, :].to_numpy()
        # print(similarity)

    # cek index data user/item yang tidak sama dengan 0
    cek_Index = np.where(ratingMatriks != 0)

    # filter user/item rating, mean centered, similarity, sesuai index data yang tidak bernilai 0
    nilai_meanCentered = np.array(meanCentered)[cek_Index]
    # print("cek nilai meanCenterd", nilai_meanCentered)
    nilai_similarity = similarity[cek_Index]
    # print("cek nilai similarity", nilai_similarity)
    # tetangga terdekat
    indexTetanggaSim = TopKTetangga(nilai_similarity, k)
    # print("cek tetangga ", indexTetanggaSim)

    # pembilang 
    pembilang = np.sum(nilai_meanCentered[indexTetanggaSim] * nilai_similarity[indexTetanggaSim])
    # print("pembilang" ,pembilang)
    # penyebut
    penyebut = np.abs(nilai_similarity[indexTetanggaSim]).sum()
    # print("pennyebuat" ,penyebut)

    # cek apakah penyebut tidak sama dengan 0
    if penyebut != 0:
        if jenis == "userBased":
            # prediksi rating
            rumusPrediksi = mean.loc[user] + (pembilang / penyebut)
        elif jenis == "itemBased":
            # rumusPrediksi rating
            rumusPrediksi = mean.loc[item] + (pembilang / penyebut)
    else:
        if jenis == "userBased":
            rumusPrediksi = mean.loc[user] + 0
        elif jenis == "itemBased":
            rumusPrediksi = mean.loc[item] + 0

    return [item, float(rumusPrediksi)] 
    
```

### PREDIKSI KESELURUHAN

```python
#keseluruhan 
def hitungKeseluruhanPrediksi(RatingMatriks, similarityFunction, mean, meanCen, k=2, jenis="userBased"):
    # Membuat DataFrame kosong untuk menyimpan prediksi
    prediksiMatriks = pd.DataFrame(index=RatingMatriks.index, columns=RatingMatriks.columns)
    
    # Iterasi untuk setiap user dan item dalam matriks rating
    for user in RatingMatriks.index:
        for item in RatingMatriks.columns:
            # Cek apakah rating untuk (user, item) adalah 0 atau tidak ada rating
            if RatingMatriks.loc[user, item] == 0:
                # Jika tidak ada rating, prediksi nilai rating menggunakan Collaborative Filtering
                prediksiMatriks.loc[user, item] = prediksiCF(RatingMatriks, similarityFunction, mean, meanCen, user, item, k, jenis)[1]
            else:
                # Jika sudah ada rating, simpan rating asli dari RatingMatriks
                prediksiMatriks.loc[user, item] = RatingMatriks.loc[user, item]

    return prediksiMatriks
```

## HYBRID

```python
def HybridFiltering(PrediksiUser, PrediksiItem, gamma=0.7):
    """
    Fungsi untuk menghitung prediksi hybrid dari prediksi user dan item.
    :param PrediksiUser: DataFrame prediksi berbasis pengguna.
    :param PrediksiItem: DataFrame prediksi berbasis item.
    :param gamma: Bobot untuk prediksi berbasis pengguna (0 <= gamma <= 1).
    :return: DataFrame hasil prediksi hybrid.
    """
    # Memastikan input adalah DataFrame
    if not isinstance(PrediksiUser, pd.DataFrame):
        raise ValueError("PrediksiUser harus berupa DataFrame")
    if not isinstance(PrediksiItem, pd.DataFrame):
        raise ValueError("PrediksiItem harus berupa DataFrame")

    # Mengonversi ke numpy array
    prediksiUser = np.array(PrediksiUser)
    prediksiItem = np.array(PrediksiItem)

    # Menghitung prediksi hybrid
    rumusHybrid = gamma * prediksiUser + (1 - gamma) * prediksiItem

    # Membuat DataFrame dari hasil prediksi hybrid dengan menggunakan index dan columns dari PrediksiUser
    dfRumusHybrid = pd.DataFrame(rumusHybrid, index=PrediksiUser.index, columns=PrediksiUser.columns)
    
    return dfRumusHybrid
```

## TOPN REKOMENDASI

```python
def TopNRecommendationPerUser(hybridPredictions, N=5):
    """
    Fungsi untuk memberikan N rekomendasi teratas untuk setiap pengguna berdasarkan prediksi rating hybrid,
    menggunakan `argsort` untuk mengurutkan.
    
    :param hybridPredictions: DataFrame dengan hasil prediksi rating hybrid untuk setiap user-item
    :param N: jumlah rekomendasi teratas yang ingin diambil untuk setiap pengguna
    :return: DataFrame dengan N rekomendasi teratas untuk setiap pengguna
    """
    topNPerUser = {}

    # Iterasi setiap user dalam DataFrame untuk mendapatkan N rekomendasi teratas
    for user in hybridPredictions.index:
        # Ambil prediksi untuk user tertentu
        userPredictions = hybridPredictions.loc[user]

        # Menggunakan argsort untuk mengurutkan prediksi rating dari yang tertinggi
        # argsort mengembalikan indeks yang diurutkan dari array
        topNIndices = userPredictions.argsort()[::-1][:N]  # Indeks untuk N teratas (dalam urutan menurun)

        # Ambil item terkait dengan indeks teratas
        topNItems = userPredictions.index[topNIndices].tolist()

        # Simpan hasil rekomendasi untuk user tersebut
        topNPerUser[user] = topNItems

    # Mengembalikan hasil rekomendasi dalam bentuk DataFrame
    dfTopNRecommendations = pd.DataFrame.from_dict(topNPerUser, orient='index', columns=[f"Rekomendasi_{i+1}" for i in range(N)])
    
    return dfTopNRecommendations



```

## EVALUASI NDCG

### DCG

```python
# evlauasi dcg

def EvaluasiDCG(GroundTruth, TopNRekomendasi, N):
    """
    Menghitung Discounted Cumulative Gain (DCG) untuk evaluasi rekomendasi.

    Args:
    - GroundTruth (pd.DataFrame): Matriks ground truth yang berisi rating sebenarnya.
    - TopNRekomendasi (pd.DataFrame): Matriks rekomendasi yang berisi rating prediksi.
    - N (int): Jumlah item teratas yang ingin dievaluasi.

    Returns:
    - float: Nilai DCG.
    """
    # Inisialisasi DCG
    dcg = 0.0

    # Iterasi melalui setiap pengguna
    for user in range(GroundTruth.shape[0]):
        # Ambil rekomendasi untuk pengguna ini
        rekomendasi = TopNRekomendasi.iloc[user, :].argsort()[::-1][:N]
        
        # Hitung DCG untuk pengguna ini
        for n in range(1, N + 1):
            # Item yang direkomendasikan
            recommended_item = rekomendasi[n - 1]
            # Rating sebenarnya dari item yang direkomendasikan untuk pengguna ini
            true_rating = GroundTruth.iloc[user, recommended_item]

            # Menghitung nilai relevansi
            relevansi = true_rating  # Relevansi bisa dianggap sama dengan rating sebenarnya

            # Menambahkan ke DCG dengan formula relevansi/log2 posisi
            dcg += relevansi / math.log2(n + 1)

    # Mengembalikan nilai DCG
    return dcg
```

### IDCG

```python
# evluasi idcg

def EvaluasiIDCG(N):
    """
    Menghitung Ideal Discounted Cumulative Gain (IDCG) untuk evaluasi rekomendasi.

    Args:
    - N (int): Jumlah item teratas yang ingin dievaluasi.

    Returns:
    - float: Nilai IDCG.
    """
    # Inisialisasi IDCG
    idcg = 0.0

    # Hitung IDCG untuk N item teratas
    for n in range(1, N + 1):
        # Menghitung nilai relevansi ideal
        relevansi_ideal = 1  # Asumsikan rating tertinggi adalah 1

        # Menambahkan ke IDCG dengan formula relevansi/log2 posisi
        idcg += sum(relevansi_ideal / math.log2(n + 1))

    # Mengembalikan nilai IDCG
    return idcg
```

### NDCG

```python
# evluasi ndcg 

def EvaluasiNDCG(GroundTruth, TopNRekomendasi, N):
    """
    Menghitung Normalized Discounted Cumulative Gain (NDCG) untuk evaluasi rekomendasi.

    Args:
    - GroundTruth (pd.DataFrame): Matriks ground truth yang berisi rating sebenarnya.
    - TopNRekomendasi (pd.DataFrame): Matriks rekomendasi yang berisi rating prediksi.


    - N (int): Jumlah item teratas yang ingin dievaluasi.
    Returns:

    - float: Nilai NDCG.
    """
    # Hitung DCG
    dcg = EvaluasiDCG(GroundTruth, TopNRekomendasi, N)

    # Hitung IDCG
    idcg = EvaluasiIDCG(N)

    # Menghitung NDCG
    rumusNDCG = dcg / idcg if idcg != 0 else 0

    return rumusNDCG
```
